```python
def get_sectors(L):
    zz_secs = [-1, 1]
    z1_secs = [-1, 1] if L % 2 == 0 else [1]
    z2_secs = [-1, 1] if L % 4 == 0 else [1]
    return [[zz, z1, z2] for zz in zz_secs for z1 in z1_secs for z2 in z2_secs]
```
